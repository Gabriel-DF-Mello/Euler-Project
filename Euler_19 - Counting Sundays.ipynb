```python
def DaysInYears(s_year,f_year,starting_day):
    
    days_in_week = ['Su','Mo','Tu','We','Th','Fr','Sa']
    index = starting_day - 1
    num_of_days = 0
    
    global_limit = 0
    limit_ja = 31
    limit_fe = 28
    limit_ap = 30
    
    months = []
    month = []
    
    
    for year in range(s_year,f_year + 1):   
        if IsLeapYear(year):
            limit_fe = 29
        else:
            limit_fe = 28
            
        for i in range(1,13):
            if (i == 11) or (i == 9) or (i == 6) or (i == 4):
                global_limit = limit_ap
            
            elif i == 2:
                global_limit = limit_fe
                
            else:
                global_limit = limit_ja
            
            for j in range(1, global_limit + 1):
                month.append(days_in_week[index])
                index = PassDay(index)
                num_of_days += 1
            months.append(month)
            print(num_of_days)
            num_of_days = 0
            month = []
    return months
```

```python
def IsLeapYear(year):
    string = str(year)
    if ((string[2] == 0) and (string[3] == 0)):
        if (year % 400 == 0):
            return True
        else:
            return False
    else:
        if (year % 4 == 0):
            return True
        else:
            return False
```

```python
def PassDay(index):
    if index == 6:
        index = 0
    else:
        index += 1
    return index
```
