```python
def IsPalindrome(n):
    string = str(n)
    len2 = round(len(string)/2)
    comparative = string[0:len2] 
    rev = comparative[::-1]
    check = comparative + rev
    
    if string == check:
        return True
    else:
        print (string)
        print (check)
        return False
```

```python
def solution():
    p = 990
    q = 999
    r = 0
    t = 0
    for p in range(990, 99, -11):
        for q in range(999, 99, -1):
            t = p * q
            if((t > r) and (IsPalindrome(t))):
                r = t
                break
            elif (t < r):
                break
    return r          
```
