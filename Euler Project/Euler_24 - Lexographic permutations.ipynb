```python
pm_list = []
```

```python

# Start by sorting the characters of the string in ascending order. This would be the first and minimal lexicographic order of the string. From here, you continue finding the next permutation until one can be found.

# 1.[The string is represented by 'str', indexed from 1..n, n being the length of the string]

# 2.Find the greatest index 'i' within the string such that str[i] < str[i+1]. If you cannot find any such index, this is the last permutation. Go to step 7.

# 3.Find the greatest index 'p' where str[i] < str[p].

# 4.Swap str[i] and str[p].

# 5.Reverse the contents of the string from str[i+1] to str[n].

# 6.The current string is the next lexicographic permutation. Go back to step 2.

# 7.End
```

```python
pm_list.clear()

string = list("0123456789")
pm_list.append("".join(string))
max_i = 0
max_p = 0

while 1:
    for i in range(1,len(string)-1):
        if (i > max_i) and (string[i+1] > string[i]):
            max_i = i

    for p in range(1,len(string)):
        if (p > max_p) and (string[max_i] < string[p]):
            max_p = p

    if (max_i == 0) and (max_p == 0):
        break

    placeholder = string[max_i]
    string[max_i] = string[max_p]
    string[max_p] = placeholder

    string[max_i+1:len(string)] = reversed(string[max_i+1:len(string)])
    
    print(max_i, ' - ', max_p)
    print(string)
    sub = "".join(string)
    pm_list.append(sub)
    
    max_i = 0
    max_p = 0
    
```
